# Random Forest Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

datos = pd.read_csv('Medicion_1234_accelGyro_modificado.csv')
datos['actividad'] = datos['actividad'].astype('category').cat.codes
X = datos.drop(['dateTime_UTC', 'actividad', 'ay', 'az', 'gx','gy','gz'], axis=1)  
Y = datos.drop(['dateTime_UTC', 'actividad', 'ax', 'az', 'gx','gy','gz'], axis=1)
Z = datos.drop(['dateTime_UTC', 'actividad', 'ax', 'ay', 'gx','gy','gz'], axis=1)
act = datos['actividad']   


datos_test = pd.read_csv('Medicion_3124_accelGyro_modificado.csv')
datos_test['actividad'] = datos_test['actividad'].astype('category').cat.codes
X_test=datos_test.drop(['dateTime_UTC','actividad', 'ay', 'az', 'gx', 'gy', 'gz'], axis=1)
Y_test = datos_test.drop(['dateTime_UTC', 'actividad', 'ax', 'az', 'gx','gy','gz'], axis=1)
Z_test = datos_test.drop(['dateTime_UTC', 'actividad', 'ax', 'ay', 'gx','gy','gz'], axis=1)
act_test=datos_test['actividad']



#Entrenar el modelo
modeloX = RandomForestClassifier()
modeloX.fit(X, act)

modeloY = RandomForestClassifier()
modeloY.fit(Y,act)

modeloZ = RandomForestClassifier()
modeloZ.fit(Z,act)

#Evaluar el modelo
X_pred = modeloX.predict(X_test)
Y_pred = modeloY.predict(Y_test)
Z_pred = modeloZ.predict(Z_test)
precisionX = accuracy_score(act_test, X_pred)
precisionY = accuracy_score(act_test, Y_pred)
precisionZ = accuracy_score(act_test, Z_pred)
print("Precisión del modelo en X:", precisionX)
print("Precisión del modelo en Y:", precisionY)
print("Precisión del modelo en Z:", precisionZ)
```
